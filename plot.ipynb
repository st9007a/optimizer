```python
import pickle
from pprint import pprint
from statistics import median

def show_result(file = 'results/GA_func1.pkl', clip = 20):

    best_so_far = []
    
    with open(file, 'rb') as p:
        res = pickle.load(p)
        
    for i in range(len(res['history'])):
        x = []
        y = []

        for history in res['history'][i]:
            x.append(history['iter'])
            y.append(history['val'])
            #y.append(math.log(history['val']))
    
        best_so_far.append(y[-1])
        
        if (clip >= len(y)):
            continue
        
        x = np.array(x)
        y = np.array(y)

        plt.scatter(x[clip:], y[clip:])
        #plt.plot(x, y, label = 'round %d' % (i + 1))

    plt.show()

    print('Average of best_so_far            : ', sum(best_so_far) / len(best_so_far))
    print('Std of best_so_far                : ', np.std(best_so_far))
    print('Mean of fitness at last iteration : ', sum(res['mean_fitness_last_iter']) / len(res['mean_fitness_last_iter']))
    print('Median of best_so_far             : ', median(best_so_far))
    
    
show_result()
```

```python
show_result('results/PSO_func1.pkl', 20)
show_result('results/PSO_func2.pkl', 20)
show_result('results/PSO_func3.pkl', 25)
show_result('results/PSO_func4.pkl', 30)
```
