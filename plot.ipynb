```python
import pickle
from pprint import pprint

def show_result(file = 'results/PSO_func1.pkl', clip = 10):

    best_so_far = []
    
    with open(file, 'rb') as p:
        res = pickle.load(p)

    for i in range(len(res['history'])):
        x = []
        y = []

        for history in res['history'][i]:
            x.append(history['iter'])
            y.append(history['val'])
            #y.append(math.log(history['val']))
    
        x = np.array(x[clip:])
        y = np.array(y[clip:])
        best_so_far.append(y[-1])

        plt.scatter(x, y)
        #plt.plot(x, y, label = 'round %d' % (i + 1))

    plt.show()
    
    print('Average of best_so_far            : ', sum(best_so_far) / len(best_so_far))
    print('Mean of fitness at last iteration : ', sum(res['mean_fitness_last_iter']) / len(res['mean_fitness_last_iter']))
    
show_result()
```
